# **Import Necessary Libraries**

---



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Loading the Dataset**

---



```python
df = pd.read_csv("pe_train.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

# **Check for NULL Values**

---



```python
df.isnull().sum()
```

```python
df.isnull().sum().value_counts()
```

```python
drop_null_cols = list()
for i in df.columns:

  if(df[i].isnull().sum() == 78543):

    print(i)
    drop_null_cols.append(i)
```

```python
df = df.drop(drop_null_cols, axis=1)
```

```python
for i in df.columns:

  if(df[i].isnull().any() == True):

    median = df[i].median()
    df[i] = df[i].fillna(median)
```

```python
df.isnull().sum().value_counts()
```

# **Splitting into Dependent & Independent Variables**

---



```python
X = df.drop(["ID_LAT_LON_YEAR_WEEK", "emission"], axis=1)
y = df["emission"]
```

# **Splitting into Training & Testing Set**

---



```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                    random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# **Model Building - Random Forest Regressor**

---



```python
from sklearn.ensemble import RandomForestRegressor
rfr_model = RandomForestRegressor()
```

```python
rfr_model.fit(X_train, y_train)
```

```python
rfr_model.score(X_test, y_test)
```

# **Model Building - XGB Regressor**

---



```python
from xgboost import XGBRegressor
xgbr_model = XGBRegressor()
```

```python
xgbr_model.fit(X_train, y_train)
```

```python
xgbr_model.score(X_test, y_test)
```

# **Testing Data**

---



```python
test_df = pd.read_csv("pe_test.csv")
test_df.head()
```

```python
test_df.shape
```

```python
test_df = test_df.drop(drop_null_cols, axis=1)
```

```python
test_df.isnull().sum().value_counts()
```

```python
for i in test_df.columns:

  if(test_df[i].isnull().any() == True):

    median = test_df[i].median()
    test_df[i] = test_df[i].fillna(median)
```

```python
test_df.isnull().sum().value_counts()
```

```python
X_subs = test_df.drop("ID_LAT_LON_YEAR_WEEK", axis=1)
```

## **Submission File - 1**

---



```python
y_preds1 = rfr_model.predict(X_subs)
```

```python
subs1_df = pd.DataFrame({
    "ID_LAT_LON_YEAR_WEEK": test_df["ID_LAT_LON_YEAR_WEEK"],
    "emission": y_preds1
})
```

```python
subs1_df.to_csv("submission1.csv", index=False)
```

## **Submission File - 2**

---



```python
y_preds2 = xgbr_model.predict(X_subs)
```

```python
subs2_df = pd.DataFrame({
    "ID_LAT_LON_YEAR_WEEK": test_df["ID_LAT_LON_YEAR_WEEK"],
    "emission": y_preds2
})
```

```python
subs2_df.to_csv("submission2.csv", index=False)
```
